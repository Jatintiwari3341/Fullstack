### 1. Why is it called as pseudo-class? ###
**Pseudo-Class:**

The term "pseudo-class" in CSS refers to a keyword that is used to define a special state of an HTML element.<br>
The prefix "pseudo-" indicates that these classes are not real or directly selectable classes in the HTML document.<br>  Instead, they represent a specific state or condition of an element, such as :hover for the hover state, :active for the active state, or :first-child for the first child of a parent element.<br>
The term "pseudo" is used to distinguish these classes from real classes defined in the HTML markup.

### 2. What are gradients in Css? ###
  
**Gradients in CSS:**

  Gradients in CSS are a way to smoothly transition between two or more colors.<br>
  CSS gradients allow you to create a blend of colors that can be applied to backgrounds, borders, or even text. 

  *There are two main types of gradients in CSS:*

**Linear Gradients:**

These gradients transition colors in a straight line, either horizontally, vertically, or at any angle specified.

Here is an example of a horizontal linear gradient and radial gradient:

```css
/* Example of a horizontal linear gradient */
background: linear-gradient(to right, red, yellow);

/* Example of a radial gradient */
background: radial-gradient(circle, red, yellow);
``` 

### 3. What are different types of transitions in Css? ###

**Transitions in CSS:**

Transitions in CSS allow you to smoothly change property values over a specified duration. They are often used to create smooth animations or effects when an element undergoes a change in state. Commonly used properties for transitions include transition-property, transition-duration, transition-timing-function, and transition-delay.<br> Here are some types of transitions:

**Color Transitions:**

```css
/* Example of a color transition */
transition: color 0.5s ease-in-out;
```
<br>

**Size Transitions:**

```css
/* Example of a size transition */
transition: width 1s ease-in-out;
```
<br>

**Opacity Transitions:**

```css
/* Example of an opacity transition */
transition: opacity 0.8s linear;
```
<br>

**Transform Transitions:**

```css
/* Example of a transform transition */
transition: transform 0.5s ease-in-out;
```

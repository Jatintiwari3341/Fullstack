### 1. What is padding and margin and when do you use them?
 - **Padding:** Padding is the space between the content of an element and its border. It is used to provide internal spacing and separate the content from the border. Padding is specified using the `padding` property in CSS.

     ```css
     .example {
       padding: 10px; /* Applies 10 pixels of padding to all sides */
     }
     ```

   - **Margin:** Margin is the space outside the border of an element. It controls the space between elements and is used to create visual separation between them. Margins are specified using the `margin` property in CSS.

     ```css
     .example {
       margin: 10px; /* Applies 10 pixels of margin to all sides */
     }
     ```

   - **When to Use Them:**
     - **Padding:** Use padding when you want to add space within an element, such as between text and a border or between an element's content and its edges.
     - **Margin:** Use margin when you want to create space around an element, separating it from other elements on the page.


### 2. What is display property and explain display inline, block, and inline-block=
   - **Display Property:** The `display` property in CSS defines how an element should be displayed. It influences the layout and behavior of the element on the page.

   - **Inline:** Elements with `display: inline;` are rendered on the same line as the content before and after them, without starting on a new line. They only take up as much width as necessary.

   - **Block:** Elements with `display: block;` start on a new line and take up the full width available. They create a "block" that pushes down any content following them.

   - **Inline-Block:** Elements with `display: inline-block;` are similar to inline elements in that they do not start on a new line, but they can have set widths and heights like block elements. They flow inline with the content.

   ```css
   .inline {
     display: inline;
   }

   .block {
     display: block;
   }

   .inline-block {
     display: inline-block;
   }
   ```

### 3. Explain min-height, min-width, max-height, and max-width in CSS?
   - **min-height:** Specifies the minimum height that an element should have. If the content is larger, the height expands to accommodate it.

   - **min-width:** Specifies the minimum width that an element should have. If the content is wider, the width expands to accommodate it.

   - **max-height:** Specifies the maximum height that an element should have. If the content is smaller, the height remains as specified, and the overflow may occur.

   - **max-width:** Specifies the maximum width that an element should have. If the content is narrower, the width remains as specified, and the overflow may occur.

   ```css
   .example {
     min-height: 100px;
     min-width: 200px;
     max-height: 300px;
     max-width: 400px;
   }
   ```

   These properties are useful for controlling the size of elements, ensuring they meet certain size constraints while adapting to varying content sizes.

### 1. Explain the Table and its properties.

A table in HTML is used to represent data in a structured way. The basic structure of an HTML table consists of the `<table>` element, which contains one or more `<tr>` (table row) elements. Each row can contain `<td>` (table data/cell) elements for regular data or `<th>` (table header) elements for header cells. Key properties include:

- **`<table>`:**
  - The container element for the entire table.
  - Common attributes: `border`, `cellspacing`, `cellpadding`, `width`, `height`.

- **`<tr>`:**
  - Represents a row in the table.
  - Common attributes: None.

- **`<td>`:**
  - Represents a data cell in a table row.
  - Common attributes: `colspan`, `rowspan`, `headers`.

- **`<th>`:**
  - Represents a header cell in a table row.
  - Common attributes: `colspan`, `rowspan`, `headers`.

Example:

```html
<table border="1" cellspacing="0" cellpadding="5">
  <tr>
    <th>Header 1</th>
    <th>Header 2</th>
  </tr>
  <tr>
    <td>Data 1</td>
    <td>Data 2</td>
  </tr>
</table>
```

### 2. Explain all Media elements and how they work.

HTML provides several media elements for embedding different types of media content:

- **`<img>`:**
  - Used to embed images.
  - Example: `<img src="image.jpg" alt="Description">`.

- **`<audio>`:**
  - Used to embed audio content.
  - Example: `<audio src="audio.mp3" controls></audio>`.

- **`<video>`:**
  - Used to embed video content.
  - Example: `<video src="video.mp4" controls></video>`.

### 3. What is Get and Post in Form?

In HTML forms, the `method` attribute is used to specify how to send form data to the server:

- **`GET`:**
  - Appends form data to the URL in the form of query parameters.
  - Suitable for non-sensitive data.
  - Visible in the browser's address bar.
  - Example: `<form action="/submit" method="get">`.

- **`POST`:**
  - Sends form data in the request body.
  - Suitable for sensitive or large amounts of data.
  - Not visible in the browser's address bar.
  - Example: `<form action="/submit" method="post">`.

### 4. List all the input types we have in forms.

HTML provides various input types that can be used within `<input>` elements in forms. Some common ones include:

- **Text Input:**
  ```html
  <input type="text" name="username">
  ```

- **Password Input:**
  ```html
  <input type="password" name="password">
  ```

- **Checkbox:**
  ```html
  <input type="checkbox" name="subscribe" value="yes">
  ```

- **Radio Button:**
  ```html
  <input type="radio" name="gender" value="male"> Male
  <input type="radio" name="gender" value="female"> Female
  ```

- **File Input:**
  ```html
  <input type="file" name="uploadFile">
  ```

- **Submit Button:**
  ```html
  <input type="submit" value="Submit">
  ```

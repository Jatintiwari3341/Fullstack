### 1. What is a Database? Explain with an example on why should we need a database.

   A database is a structured collection of data that is organized and stored in a computer system. It is designed to manage and manipulate data efficiently. In a database, data is organized into tables, where each table consists of rows and columns. Each row represents a record, and each column represents a field or attribute. For example, consider a university database where there are tables for students, courses, and grades. The students' table might have columns like student ID, name, and date of birth.

### 3. What is DBMS? What was the need for DBMS?

   File-based storage systems involve storing data in files, often in a hierarchical or tree-like structure. Each department or user maintains its own set of files. Major challenges of a file-based system include:
   - **Data Redundancy:** Same data may be duplicated in multiple files, leading to redundancy.
   - **Data Inconsistency:** Updates in one file may not reflect in others, causing inconsistency.
   - **Limited Data Sharing:** Difficulty in sharing data across different departments or users.
   - **Data Dependence:** Programs are highly dependent on the file structure, making it inflexible.
   - **Security Issues:** Limited access control and security measures.

   A DBMS is software that provides an interface to interact with databases. It facilitates the creation, maintenance, and manipulation of databases. The need for DBMS arises due to:
   - **Data Independence:** Separates the database structure from the application programs.

   - **Concurrent Access:** Allows multiple users to access and modify data simultaneously.
   - **Data Security:** Provides access control and authentication measures.
   - **Data Abstraction:** Presents a simplified view of the data to the users.

### 6. What is the significance of Data Modelling and explain the types of data modeling

   - **Based on Data Model:**
     - *Relational DBMS (RDBMS):* Organizes data into tables with relationships.
     - *NoSQL DBMS:* Suitable for unstructured or semi-structured data.
   - **Based on Number of Users:**
     - *Single-user DBMS:* Designed for a single user.
     - *Multi-user DBMS:* Supports multiple users concurrently.
   - **Based on Data Access Method:**
     - *Centralized DBMS:* Data is stored at a central location.
     - *Distributed DBMS:* Data is distributed across different locations.
   - **Based on Database Structure:**
     - *Hierarchical DBMS:* Organizes data in a tree-like structure.
     - *Network DBMS:* Utilizes a network-like data model.
     
### 7. Explain 3 schema architecture along with its advantages.
   - **Significance:**
     - *Simplifies Complexity:* Represents complex systems in a structured manner.
     - *Communication:* Provides a common language for communication between stakeholders.
     - *Blueprint for Database:* Serves as a blueprint for designing the database.
   - **Types:**
     - *Conceptual Data Model:* Abstract representation of entities and their relationships.
     - *Logical Data Model:* Describes data in terms of tables, relationships, and constraints.
     - *Physical Data Model:* Specifies how data will be stored, accessed, and implemented.

### 7. **Three Schema Architecture and Advantages:**
   - **Three Schema Architecture:**
     - *User Schema (View):* Describes how users perceive the data.
     - *Logical Schema (Conceptual):* Represents the logical structure of the entire database.
     - *Physical Schema:* Defines how the data is physically stored and accessed.
   - **Advantages:**
     - *Data Independence:* Changes in one schema do not affect others, providing abstraction layers.
     - *Security:* Access to sensitive data can be controlled at the schema level.
     - *Flexibility:* Modifications to the database structure can be made without affecting user views.

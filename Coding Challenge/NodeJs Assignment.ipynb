Certainly! You can use the `os` module in Node.js to obtain information about the operating system. Here's a simple script that prints the OS name and version:

```js
// getOSInfo.js
const os = require('os');

// Get the operating system name
const osName = os.type();

// Get the operating system release version
const osRelease = os.release();

// Print the information to the console
console.log(`Operating system name: ${osName}`);
console.log(`OS release: ${osRelease}`);
```

### 9. While working with the events module, one interesting observation is that when an event is created and called, the associated event handler is triggered. However, what happens if we remove an event and then try to call it? In this coding challenge let's create an event handler and call it. Later let's remove the event handler and observe what happens when we call it.

**Step 7: Create an HTTP Server**

Create a file, for example, `httpServer.js`, to create a simple HTTP server that responds with the desired text:

```js
// httpServer.js
const http = require('http');

// Create an HTTP server
const server = http.createServer((req, res) => {
  // Set the content type
  res.setHeader('Content-Type', 'text/plain');

  // Send the message as a response
  res.end('I Am Happy To Learn Full Stack Web Development From PW Skills!');
});

// Set the port for the server to listen on
const port = 3000;
server.listen(port, () => {
  console.log(`Server is listening on port ${port}`);
});
```

Run the server using the following command:

```bash
node httpServer.js
```

Visit `http://localhost:3000` in your web browser, and you should see the message displayed.

**Step 8: Simulate a Subscription Feature**

Create a file, for example, `subscribeEvent.js`, to simulate the subscription feature using the events module:

```js
// subscribeEvent.js
const events = require('events');

// Create an EventEmitter
const eventEmitter = new events.EventEmitter();

// Subscribe event handler
eventEmitter.on('subscribe', (channel) => {
  console.log(`Thanks For Subscribing to ${channel}`);
});

// Trigger the 'subscribe' event
eventEmitter.emit('subscribe', 'College Wallah');
```

Run the script using the following command:

```bash
node subscribeEvent.js
```

This will print the message to the console:

```
Thanks For Subscribing to College Wallah
```

**Step 9: Remove an Event Handler**

Modify the `subscribeEvent.js` file to remove the event handler after it's triggered:

```js
// subscribeEvent.js
const events = require('events');

// Create an EventEmitter
const eventEmitter = new events.EventEmitter();

// Subscribe event handler
const subscribeHandler = (channel) => {
  console.log(`Thanks For Subscribing to ${channel}`);
};

// Add the event handler
eventEmitter.on('subscribe', subscribeHandler);

// Trigger the 'subscribe' event
eventEmitter.emit('subscribe', 'College Wallah');

// Remove the event handler
eventEmitter.removeListener('subscribe', subscribeHandler);

// Trigger the 'subscribe' event again
eventEmitter.emit('subscribe', 'College Wallah');
```

Run the script using the following command:

```bash
node subscribeEvent.js
```

You'll observe that after removing the event handler, attempting to trigger the 'subscribe' event again won't produce any output, indicating that the event handler has been successfully removed.

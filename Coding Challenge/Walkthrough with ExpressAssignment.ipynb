



### Q.1 Build a Server Using Http Module NodeJS with api endpoints for Getting Products Data

    '/' ---send msg as "Welcome to Men & Women Dummy Data"

⚫'/men'-- send 10 products data of men

⚫'/women'-- send 10 products data of women

•'/other'-- send response as page not found

Hint: You can use any data, send data in json format

```js
// server.js
const http = require('http');

// Sample data for men's and women's products
const menProducts = [...]; // Array of 10 men's products
const womenProducts = [...]; // Array of 10 women's products

// Create an HTTP server
const server = http.createServer((req, res) => {
  // Set the content type to JSON
  res.setHeader('Content-Type', 'application/json');

  // Handle different API endpoints
  if (req.url === '/') {
    // Welcome message for the root endpoint
    res.end(JSON.stringify({ message: 'Welcome to Men & Women Dummy Data' }));
  } else if (req.url === '/men') {
    // Send 10 products data for men
    res.end(JSON.stringify({ products: menProducts }));
  } else if (req.url === '/women') {
    // Send 10 products data for women
    res.end(JSON.stringify({ products: womenProducts }));
  } else {
    // Page not found for other endpoints
    res.statusCode = 404;
    res.end(JSON.stringify({ message: 'Page not found' }));
  }
});

// Set the port for the server to listen on
const port = 8081;
server.listen(port, () => {
  console.log(`Server is listening on port ${port}`);
});
```

In this example, replace `[...]` with your actual product data arrays for men and women.

To run the server, save this script in a file (e.g., `server.js`) and execute it using the following command:

```bash
node server.js
```

Now you can access the API endpoints:

- `http://localhost:8081/` - Welcome message
- `http://localhost:8081/men` - Products data for men
- `http://localhost:8081/women` - Products data for women
- Any other endpoint will return a "Page not found" response with a status code of 404.

### Q.2 Create a Backend For Counter Web App with Api Endpoint for Managing Counter Using Express

'/'- send counter data as [counter.counter]

`/increment` - increment counter by I and send in response latest data as (counter.counter] ⚫`/decrement`-- decrement counter by 1 and send in response latest data as (counter.counter]

Note: Using Express is Mandatory for this Question

Certainly! Below is an example of creating a simple Express backend for a counter web app with API endpoints for managing the counter:

```js
// counterApp.js
const express = require('express');
const app = express();
const PORT = 3000;

// Initial counter value
let counter = 0;

// Middleware to parse JSON requests
app.use(express.json());

// API endpoint to get current counter value
app.get('/', (req, res) => {
  res.json({ counter: counter });
});

// API endpoint to increment the counter
app.post('/increment', (req, res) => {
  const incrementBy = req.body.incrementBy || 1; // Default increment by 1

  counter += incrementBy;
  res.json({ counter: counter });
});

// API endpoint to decrement the counter
app.post('/decrement', (req, res) => {
  counter--;
  res.json({ counter: counter });
});

// Start the server
app.listen(PORT, () => {
  console.log(`Server is running on http://localhost:${PORT}`);
});
```

In this example, we're using Express, and there are three API endpoints:

1. `GET /` - Returns the current counter value.
2. `POST /increment` - Increments the counter by the specified value (default is 1) and returns the updated counter value.
3. `POST /decrement` - Decrements the counter by 1 and returns the updated counter value.

To run the server:

1. Make sure you have Node.js and npm installed.
2. Save the above code in a file (e.g., `counterApp.js`).
3. Open your terminal and navigate to the folder where the file is saved.
4. Run the following commands:

```bash
npm init -y  # Initialize a new npm project (if not already done)
npm install express  # Install Express
node counterApp.js  # Run the server
```

Now, you can access the following endpoints:

- `http://localhost:3000/` - Returns the current counter value.
- `http://localhost:3000/increment` - Increments the counter and returns the updated value.
- `http://localhost:3000/decrement` - Decrements the counter and returns the updated value.

### Q.3 Create a Basic Server with Different Routes using Express

`/`-send response as (msg:'I am homepage]

⚫`/about`-send response as (msg:'I am about page]

⚫`/contact`-send response as [email: support@pwskills.com"}

Certainly! Below is an example of creating a basic Express server with different routes:

```javascript
// server.js
const express = require('express');
const app = express();
const PORT = 3000;

// Middleware to parse JSON requests
app.use(express.json());

// Route for the homepage
app.get('/', (req, res) => {
  res.json({ msg: 'I am homepage' });
});

// Route for the about page
app.get('/about', (req, res) => {
  res.json({ msg: 'I am about page' });
});

// Route for the contact page
app.get('/contact', (req, res) => {
  res.json({ email: 'support@pwskills.com' });
});

// Start the server
app.listen(PORT, () => {
  console.log(`Server is running on http://localhost:${PORT}`);
});
```

In this example, we have three routes:

1. `GET /` - Returns the response `{ msg: 'I am homepage' }`.
2. `GET /about` - Returns the response `{ msg: 'I am about page' }`.
3. `GET /contact` - Returns the response `{ email: 'support@pwskills.com' }`.

To run the server:

1. Make sure you have Node.js and npm installed.

3. Open your terminal and navigate to the folder where the file is saved.
4. Run the following commands:

```bash
npm init -y  # Initialize a new npm project (if not already done)
npm install express  # Install Express
node server.js  # Run the server
```

### Q.4 Build a Server that Generates Random Number Using Express

•'/random'--send random number in response {random.10}

Certainly! You can create an Express server that generates a random number and sends it in the response. Below is an example:

```javascript
// server.js
const express = require('express');
const app = express();
const PORT = 8081;

// Middleware to parse JSON requests
app.use(express.json());

// Route for generating a random number
app.get('/random', (req, res) => {
  const randomNumber = Math.floor(Math.random() * 10) + 1; // Generate a random number between 1 and 10
  res.json({ random: randomNumber });
});

// Start the server
app.listen(PORT, () => {
  console.log(`Server is running on http://localhost:${PORT}`);
});
```

In this example:

- The `/random` route generates a random number between 1 and 10 and sends it in the response.

To run the server:

1. Make sure you have Node.js and npm installed.
2. Save the above code in a file (e.g., `server.js`).
3. Open your terminal and navigate to the folder where the file is saved.
4. Run the following commands:

```bash
npm init -y  # Initialize a new npm project (if not already done)
npm install express  # Install Express
node server.js  # Run the server
```

Now, you can access the following endpoint:

- `http://localhost:8081/random` - Returns a JSON response with a random number between 1 and 10.

You can test this endpoint using tools like Postman or by making HTTP requests from your browser or any other client.

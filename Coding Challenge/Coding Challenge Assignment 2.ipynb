### 1. Count the occurrences.

You are building a word count generator that will take a large string of text as input and output the words and the number of times they are present in the string. Your task is to write a function that can count the occurrences of each word in the string and return a Map containing each word's count. For simplicity, you can assume that the string only contains alphabetic characters and spaces.


```javascript
function wordCount(text) {
  const words = text.split(' ');
  const wordMap = new Map();

  words.forEach(word => {
    const normalizedWord = word.toLowerCase();
    const count = wordMap.has(normalizedWord) ? wordMap.get(normalizedWord) + 1 : 1;
    wordMap.set(normalizedWord, count);
  });

  return wordMap;
}

// Example usage:
const text = "This is a sample text. This text will be used for word count.";
const result = wordCount(text);
console.log(result);
```

### 3. Swap the values.

You are working on a project that requires you to swap the values of two variables without using a temporary variable. You docide to write a function that takes two variables as input and swaps their values using destructuring assignment with an array. The function should take the two variables as arguments, destructure them into an array, and then swop their positions within the array. Finally, the function should return an array with the swapped values. For example, if the function is called with the arguments x = 5 and y = 10 It should return an array [10, 5] with x now equal to 10 and y equal to 5.


```javascript
function swapValues(x, y) {
  [x, y] = [y, x];
  return [x, y];
}

// Example usage:
let a = 5, b = 10;
const swappedValues = swapValues(a, b);
console.log(swappedValues); // Output: [10, 5]
```

### 5. Min and max values.

You are building a program that needs to find the maximum and minimum values in an array of numbers. decide to write a function that can accomplish this task by taking an array of numbers as input and using ta spread operator with the Math methods max() and min(). The function should return an object with two properties, max, and min, representing the maximum and minimum values in the input array, respectively. F example, if the input array is [5, 2, 7, 1, 9], the function should return an object { max: 9, min: 1].


```javascript
function findMinMax(numbers) {
  return {
    max: Math.max(...numbers),
    min: Math.min(...numbers),
  };
}

// Example usage:
const inputArray = [5, 2, 7, 1, 9];
const result = findMinMax(inputArray);
console.log(result); // Output: { max: 9, min: 1 }
```

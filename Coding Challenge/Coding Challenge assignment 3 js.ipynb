### 3. Write a function that takes two numbers as arguments and returns their sum. Declare a variable Inside the function using the var keyword and log its value to the console before it is assigned a value to demonstrate variable hoisting.

```javascript
function sumWithVar(a, b) {
  console.log(x); // Outputs: undefined
  var x = a + b;
  console.log(x);
}

sumWithVar(2, 3);
```

### 4. Declare three variables, one using let, one using var, and one using const, all inside a block scope. Assign thern values and log their values to the console before and after they are declared to demonstrate variable hoisting.


```javascript
{
  console.log(letVar); // Outputs: ReferenceError: Cannot access 'letVar' before initialization
  console.log(varVar); // Outputs: undefined
  console.log(constVar); // Outputs: ReferenceError: Cannot access 'constVar' before initialization

  let letVar = "Let Variable";
  var varVar = "Var Variable";
  const constVar = "Const Variable";
}

console.log(varVar); // Outputs: Var Variable
console.log(constVar); // Outputs: ReferenceError: constVar is not defined
```

### 5. Declare a variable using let inside a block scope and attempt to log its value to the console before it is assigned a value to demonstrate the temporal dead zone.

```javascript
{
  console.log(letVar); // Outputs: ReferenceError: Cannot access 'letVar' before initialization
  let letVar = "Let Variable in Temporal Dead Zone";
}
```

These examples demonstrate various aspects of hoisting and the temporal dead zone in JavaScript. Note that trying to access variables before they are declared can result in ReferenceError in some cases.
